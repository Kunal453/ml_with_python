# distplot

### the displot shows the distribution of a unvairate set of observation

```python
tips['total_bill']
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],bins=244) #bins desc's the number of bar present in graph
```

## jointplot

##### jointplot allows you to match up two distpots for bivariate data. with you rchoce of what kind parameter to compare with

```python
sns.jointplot(x = 'total_bill', y = 'tip', data=tips,kind='scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data=tips,kind='hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data=tips,kind='reg')
```
